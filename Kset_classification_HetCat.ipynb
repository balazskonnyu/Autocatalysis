```python
import torch.utils.data as data
class KDataset(data.Dataset):

    def __init__(self, df, n_labels=3):
        """
        Inputs:
            df - pandas dataframe containing the data
        """
        super().__init__()
        self.n_labels = n_labels
        self.data = df
        self.ks = df.values[:, 1:38].astype(np.float32)
        self.labels = df.values[:, 0].astype(int)-1

    def __len__(self):
        # Number of data points
        return len(self.data)

    def __getitem__(self, i):
        # Return the i-th data point of the dataset
        k_values = self.ks[i]
        label = self.labels[i]
        labels = torch.tensor([1 if i==label else 0 for i in range(self.n_labels)])
        return k_values, labels
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SimpleClassifier(nn.Module):

    def __init__(self, n_inputs, n_hidden, n_outputs):
        super().__init__()
        # Initialize the modules we need to build the network
        self.linear1 = nn.Linear(n_inputs, n_hidden)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(n_hidden, n_outputs)

    def forward(self, x):
        # Perform the calculation of the model to determine the prediction
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        x = torch.sigmoid(x)
        return x
```

```python
model = SimpleClassifier(n_inputs=37, n_hidden=32, n_outputs=3)
print(model)
```

```python
for name, param in model.named_parameters():
    print(f"Parameter {name}, shape {param.shape}")
```

### Optimization

After defining the model and the dataset, it is time to prepare the optimization of the model. During training, we will perform the following steps:

1. Get a batch from the data loader
2. Obtain the predictions from the model for the batch
3. Calculate the loss based on the difference between predictions and labels
4. Backpropagation: calculate the gradients for every parameter with respect to the loss
5. Update the parameters of the model in the direction of the gradients

```python
loss_fn = nn.BCELoss()
```

```python
# Input to the optimizer are the parameters of the model: model.parameters()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
# Push model to device. Has to be only done once
device = 'cuda:0'
model.to(device)
```

```python
def train_model(model, optimizer, train_loader, val_loader, loss_module, decision_threshold=0.5, num_epochs=100):

    best_validation_acc = 0

    for epoch in tqdm(range(num_epochs)):
        # training
        model.train()
        for train_input, train_labels in train_loader:

            ## Step 1: Move input data to device (only necessary if we use GPU)
            train_input = train_input.to(device)
            train_labels = train_labels.to(device)

            ## Step 2: Run the model on the input data
            train_preds = model(train_input)
            train_preds = train_preds.squeeze(dim=1) # Output is [Batch size, 1], but we want [Batch size]

            ## Step 3: Calculate the loss
            train_loss = loss_fn(train_preds, train_labels.float())

            ## Step 4: Perform backpropagation
            # Before calculating the gradients, we need to ensure that they are all zero.
            # The gradients would not be overwritten, but actually added to the existing ones.
            optimizer.zero_grad()
            # Perform backpropagation
            train_loss.backward()

            ## Step 5: Update the parameters
            optimizer.step()

        # validation
        model.eval()
        for val_input, val_labels in val_loader:
            val_losses = []
            n_correct_preds = 0
            n_all_preds = 0

            val_input = val_input.to(device)
            val_labels = val_labels.to(device)

            val_preds = model(val_input)
            class_preds = torch.argmax(val_preds, dim=1)
            val_loss = loss_fn(val_preds, val_labels.float())
            val_losses.append(val_loss.item())
            n_correct_preds += sum(class_preds==torch.argmax(val_labels, dim=1))
            n_all_preds += len(val_labels)

        current_acc = n_correct_preds / n_all_preds
        print(f'Average validation loss: {np.mean(val_losses)}')
        print(f'Average accuracy: {current_acc}')
        if current_acc > best_validation_acc:
            torch.save(model.state_dict(), "kset_model_HetCat2_NEW.pth")
            print('Model weights saved')
            best_validation_acc = current_acc
```

```python
train_model(model, optimizer, train_loader, val_loader, loss_fn, num_epochs=10)
```
